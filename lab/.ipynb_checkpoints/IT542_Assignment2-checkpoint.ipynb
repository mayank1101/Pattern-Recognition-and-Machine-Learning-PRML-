```python
import numpy as np
import random
from scipy.stats import norm
import math


np.random.seed(123)
sampleG1 = np.array(np.random.normal(10,math.sqrt(20), 200))
sampleG2 = np.array(np.random.normal(20,math.sqrt(25), 200))

#compute sample mean and sample variance
mu1, var1 = np.mean(sampleG1), np.var(sampleG1)
mu2, var2 = np.mean(sampleG2), np.var(sampleG2)

# draw 40 random samples
random_sample = []
for i in range(40):
    random_sample.append(random.randint(5,21))
np_random_sample = np.array(random_sample)
#print(np_random_sample)

P1, P2, P3 = (0.5, 0.5), (0.3, 0.7), (0.7, 0.3)
print("Classification for P1(0.5, 0.5)")
for point in np_random_sample:
    pdf1, pdf2 = norm(mu1, var1).pdf(point), norm(mu2, var2).pdf(point)
    if((P1[0] * pdf1) >= (P1[1] * pdf2)):
        print("Class A")
    else:
        print("Class B")
        
print("Classification for P2(0.3, 0.7)")
for point in np_random_sample:
    pdf1, pdf2 = norm(mu1, var1).pdf(point), norm(mu2, var2).pdf(point)
    if((P2[0] * pdf1) >= (P2[1] * pdf2)):
        print("Class A")
    else:
        print("Class B")

print("Classification for P2(0.7, 0.3)")
for point in np_random_sample:
    pdf1, pdf2 = norm(mu1, var1).pdf(point), norm(mu2, var2).pdf(point)
    if((P3[0] * pdf1) >= (P3[1] * pdf2)):
        print("Class A")
    else:
        print("Class B")
```
