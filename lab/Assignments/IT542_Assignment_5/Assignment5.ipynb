```python
points = np.vstack(X) #Stacks sequence of arrays vertically
print(points)
```

```python
def centroid_initialization(points, k):
    centroids = points.copy()
    np.random.shuffle(centroids)
    return centroids[:k]

def closest_centroid(points, centroids):
    d = np.sqrt(((points - centroids[:, np.newaxis])**2).sum(axis=2))
    return np.argmin(d, axis=0)
```

```python
c = centroid_initialization(points,3)
new_min= closest_centroid(points, c)
```
