```python
'''Question1 : Draw 1000 samples from p(x) = P1.p1(x) + P2.p2(x)
Where, P1 = 0.6, P2 = 0.4, p1(x) ~ N(5,10) and p2(x) ~ N(10,15).
Draw a histogram for the data.'''


import numpy as np
import math


P1, P2 = 0.6, 0.4
sample1 = np.random.normal(5, math.sqrt(10), 600)
sample2 = np.random.normal(10, math.sqrt(15), 400)
sample = np.concatenate((sample1, sample2))
print(len(sample))
```

```python
import matplotlib.pyplot as plt


plt.hist(sample, bins =10, rwidth=0.85, color='green', alpha=0.7)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show
plt.title('Histogram')
```

```python
import numpy as np
import matplotlib.pyplot as plt
sample1 = np.random.normal(5,1,100)
sample2 = np.random.normal(-5,1,100)
data=sample1
data= np.append(data,sample2)

print(np.shape(data))
N=200
print(data)
```

```python
def phi(x):
 return (np.exp(-0.5*(x**2))/(np.sqrt(2*np.pi)))  # Normal Density-> Mean =0, var=1
           
def tpdf(x):
 return (phi(x-5)*0.5) +  (phi(x+5)*0.5) # True Density-> Mean 10,-10, var=1



def kernel(x,h):
    return np.mean(phi((x-data)/h)/h) #  Kernel Density
    
def kpdf(x,h):
    temp=[]
    for i in range(len(x)):
        temp.append(kernel(x[i],h))
        
    return temp   
```

```python
plt.plot(sample1,'*')
plt.plot(sample2,'o')
plt.plot(data[0],'.')
plt.plot(data[1],'.')
plt.show()
```
